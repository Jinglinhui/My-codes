```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
```

```python
customer_example=pd.read_csv('/Users/jinglinhui/Desktop/Data_analytics_project/week2_2/customer_example.csv')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data= scaler.fit_transform(customer_example)
```

## Use Dendrogram to choose the number of clusters

```python
dendrogram = sch.dendrogram(sch.linkage(scaled_data, method  = "ward"))
plt.title('Dendrogram')
plt.xlabel('Satisfaction')
plt.ylabel('Euclidean distances')
plt.show()
```

### The x-axis consists of the Satisfactions and y-axis consists of the Euclidean distance between the clusters. To determine the optimal number of clusters from this diagram, we need to look for the largest distance that we can vertically without crossing any horizontal line and this one is the orange framed line on the above diagram. Cluster number will be 5 for this dataset

```python
from sklearn.cluster import AgglomerativeClustering 
h_clustering= AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage ='ward')
```

```python
data_hclustering=h_clustering.fit_predict(scaled_data)
```

```python
from matplotlib.pyplot import figure
figure(figsize=(9,5), dpi=80)
label = h_clustering.fit_predict(scaled_data)
u_labels = np.unique(label)
for i in u_labels:
    plt.scatter(scaled_data[label == i , 0] , scaled_data[label == i , 1] , label = i)
plt.legend()
plt.show()
```
